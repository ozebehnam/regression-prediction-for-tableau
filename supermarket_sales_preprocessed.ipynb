```python
import pandas as pd
```

```python
df = pd.read_csv('supermarket_sales.csv')
```

```python
df.head()
```

# Create Dummies

```python
def dummies_var(df, column):
    name = column+"_name"
    name = pd.get_dummies(df[column])
    df = pd.concat([df, name], axis='columns')
    return df
```

```python
new_df = df.copy()

new_df = dummies_var(new_df,'Branch')

new_df = dummies_var(new_df,'City')

new_df = dummies_var(new_df,'Customer type')

new_df = dummies_var(new_df,'Gender')

new_df = dummies_var(new_df,'Product line')

new_df = dummies_var(new_df,'Payment')
new_df.head()
```

# Remove converted columns

```python
cleaned_df = new_df.copy()

cleaned_df.drop('Branch', inplace=True, axis=1)

cleaned_df.drop('City', inplace=True, axis=1)

cleaned_df.drop('Product line', inplace=True, axis=1)

cleaned_df.drop('Payment', inplace=True, axis=1)

cleaned_df.drop('Customer type', inplace=True, axis=1)

cleaned_df.drop('Gender', inplace=True, axis=1)

cleaned_df.head()
```

# Usefull features and change Date Time format

```python
cleaned_df = cleaned_df.iloc[:,1:]
cleaned_df
```

```python
import datetime
```

```python
cleaned_df['Date'] = pd.to_datetime(cleaned_df['Date'], format = '%m/%d/%Y')

cleaned_df['Day of week'] = cleaned_df['Date'].dt.day_of_week
cleaned_df = cleaned_df.drop(['Date', 'Time'], axis=1)
cleaned_df
```

```python
cleaned_df.to_csv('preprocessed.csv')
```
