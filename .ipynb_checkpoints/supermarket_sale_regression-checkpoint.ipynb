```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed = pd.read_csv('preprocessed.csv')
```

```python
data_preprocessed.head()
```

```python
data_preprocessed['Rating'].median()
```

```python
targets = np.where(data_preprocessed['Rating'] > 
                  data_preprocessed['Rating'].median(),1 ,0)

data_preprocessed['Class'] = targets

data_preprocessed.head()
```

# Add target class to dataframe

```python
labled_df = data_preprocessed.copy()
```

# Scale Data

```python
df_columns = labled_df.columns.view()

df_columns
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()

df_scaled = scaler.fit_transform(labled_df.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs',
       'gross margin percentage', 'gross income', 'Rating', 'A', 'B', 'C',
       'Mandalay', 'Naypyitaw', 'Yangon', 'Member', 'Normal', 'Female', 'Male',
       'Electronic accessories', 'Fashion accessories', 'Food and beverages',
       'Health and beauty', 'Home and lifestyle', 'Sports and travel', 'Cash',
       'Credit card', 'Ewallet', 'Day of week', 'Class'])


df_scaled_ready = df_scaled.copy()

target = df_scaled_ready['Class']

df_scaled_ready.drop(['Class'], axis=1)
```

# Train and test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_scaled_ready, targets, train_size= 0.8, random_state= 20)

print(x_train.shape, y_train.shape)

print(x_test.shape, y_test.shape)
```

# Train regression machine

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

reg = LogisticRegression()

reg.fit(x_train, y_train)

reg.score(x_train, y_train)
```

# Reverse MinMaxScaler effect

```python
model_outputs = reg.predict(x_train)

df_final = x_train

final = pd.DataFrame(scaler.inverse_transform(x_train))

final.columns = ['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs',
       'gross margin percentage', 'gross income', 'Rating', 'A', 'B', 'C',
       'Mandalay', 'Naypyitaw', 'Yangon', 'Member', 'Normal', 'Female', 'Male',
       'Electronic accessories', 'Fashion accessories', 'Food and beverages',
       'Health and beauty', 'Home and lifestyle', 'Sports and travel', 'Cash',
       'Credit card', 'Ewallet', 'Day of week', 'Class']

final['prediction'] = model_outputs

final
```

```python
final.to_csv('final_prediction.csv')
```
